```python
import cv2
import os

vidcap=cv2.VideoCapture('/content/drive/MyDrive/face_project/kkk.mp4')

cnt=0
os.mkdir('/content/drive/MyDrive/face_project/kkk/')

while(vidcap.isOpened()):
  ret, image = vidcap.read()
  if ret == False:
    break

  if(int(vidcap.get(1)) % 100 == 0): # 25프레임당 1장씩 저장 == 1초
    if cnt % 10 ==0:
      print('Save img number:' + str(int(cnt)))
    cv2.imwrite('/content/drive/MyDrive/face_project/kkk/gameyong_%d.jpg'%cnt, image)
    cnt+=1

vidcap.release()
#cv2.destroyAllWindows()
```

```python
from IPython.display import Image, clear_output
from glob import glob

img_list = glob('/content/drive/MyDrive/face_project/kkk/*.jpg')
print(len(img_list))
txt_list = glob('/content/drive/MyDrive/face_project/kkk/txt/*.txt')
print(len(txt_list))
```

```python
# 파일 이동

import os
import shutil

path = input('이동 할 파일 경로') + '/'
files = os.listdir(path)

new_path = input('이동 시킬 파일 경로') + '/'
if not os.path.exists(new_path):
    os.mkdir(new_path)
    

# move
extend = input('파일 확장자')
for file in files:
    if extend in file:
        shutil.move(path + file, new_path + file)
        print('{} has been moved to new folder!'.format(file))

# # copy
# for file in files:
#     if 'pdf' in file:
#         shutil.copy(path + file, new_path + file)
#         print('{} has been copied in new folder!'.format(file))
```
