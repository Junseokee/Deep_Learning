```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import date, timedelta
import time

url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=pnt&date='
YY = int(input("기준 년도를 입력하세요 : "))
mm = int(input("기준 월을 입력하세요 : "))
dd = int(input("기준 일을 입력하세요 : "))
period = int(input('크롤링 할 기간을 입력 해 주세요: '))
```

```python
start_date = date(YY,mm,dd)
print(start_date)
timegap = timedelta(days=period)
print(timegap)
start_date = start_date - timegap
print(start_date)
start_date = start_date.strftime('%Y-%m-%d')
print(start_date)
range_date = pd.date_range(start_date, periods=period, freq='D')
```

```python
range_date
```

```python
tit_lst = []
point_lst = []
date_lst = []
for today in range_date:
    # today.strftime() 형식변환
    html = requests.get(url+today.strftime('%Y%m%d'))
    time.sleep(0.1)
    soup = BeautifulSoup(html.text, "html.parser")
    
    #해당 페이지에서 영화 제목 추출
    tit5_tag = soup.find_all('div',class_='tit5')
    #해당 페이지에서 영화 평점 추출
    point_tag = soup.find_all('td',class_='point')
    for num in range(len(tit5_tag)):
        #tit_5태그의 num번째 요소에 접근해 영화 제목과 평점 추출
        title = tit5_tag[num].find('a').text
        tit_lst.append(title)
        
        point = point_tag[num].text
        point_lst.append(point)
        
        # 날짜
        date_lst.append(today.strftime('%Y-%m-%d'))
#데이터 프레임 저장
movie_df = pd.DataFrame({"date" : date_lst,
                         "title" : tit_lst,
                        "point" : point_lst})
movie_df.to_csv('C:/Users/sec/Desktop/Junseok/네이버 영화 평점 크롤링.csv', encoding='utf-8')
```

```python
start_date = date(YY,mm,dd)
print(start_date)
timegap = timedelta(days=period)
print(timegap)
start_date = start_date - timegap
print(start_date)
start_date = start_date.strftime('%Y-%m-%d')
print(start_date)
range_date = pd.date_range(start_date, periods=period, freq='D')
```
